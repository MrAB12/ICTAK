```python
import pandas as pd
import seaborn as sns
```

1.Load Dataset into python

```python
titanic_df = sns.load_dataset('titanic')
titanic_train = pd.read_csv('/content/titanic_dataset.csv')
titanic_df.head()
```

2. Make ‘PassengerId’ as the index column

```python
titanic_train.set_index('PassengerId', inplace=True)
titanic_train.head(10)
```

4. Fill in all the missing values present in all the columns in the dataset

```python
for col in titanic_train.select_dtypes(include=['int64', 'float64']).columns:
    titanic_train[col].fillna(titanic_train[col].mean(), inplace=True)
    print(titanic_train.head())
```

5. Check and handle outliers in at least 3 columns in the dataset

```python
from scipy import stats
import numpy as np

columns_to_check = ['Age', 'Fare', 'SibSp']

for col in columns_to_check:
    z_scores = stats.zscore(titanic_train[col])
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < 3)
    new_data = titanic_train[filtered_entries]
    print(f"{col}: {np.sum(~filtered_entries)} outliers removed.")
```

6. Do min max scaling on the feature set (Take ‘Survived’ as target)


```python
from sklearn.preprocessing import MinMaxScaler

numerical_features = titanic_train.select_dtypes(include=['int64', 'float64'])

features = numerical_features.drop('Survived', axis=1)
target = titanic_train['Survived']

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaled_titanic_train = pd.DataFrame(scaled_features, columns=features.columns)

scaled_titanic_train['Survived'] = target
print(scaled_titanic_train.head())
```
